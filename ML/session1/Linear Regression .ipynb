```python
import numpy as np
from matplotlib import pyplot as plt

np.random.seed(42)

x = np.linspace(-2, 2, 31)
f = lambda x: 3 * x + 17
y = f(x) + np.random.randn(len(x))

plt.plot(x, f(x), '--', color = 'red', label = 'Ground Truth')
plt.plot(x, y, 'o', color = 'blue', label = 'Data')
plt.title('Linear Regression Data')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

$\omega_{i}^{(t + 1)} = \omega_{i}^{(t)} - \alpha \times \frac{\partial{J}}{\partial{\omega_{i}^{(t)}}}$

$b^{(t + 1)} = b^{(t)} - \alpha \times \frac{\partial{J}}{\partial{b^{(t)}}}$

$\frac{\partial{J}}{\partial{\omega_{i}}} = \frac{-1}{N} \sum_{j = 1}^{N} x_{ij} e_j$

$\frac{\partial{J}}{\partial{b}} = \frac{-1}{N} \sum_{j = 1}^{N} e_j$

$e_j = y_j - \hat{y_j}$

```python
def display(y_pred):
    plt.plot(x, f(x), '--', color = 'red', label = 'Ground Truth')
    plt.plot(x, y, 'o', color = 'blue', label = 'Data')
    plt.plot(x, y_pred, '-', color = 'green', label = 'Predicted')
    plt.title('Linear Regression Example')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()
    plt.show()
```

```python
alpha = 0.01
r = 1e-4
n_iter = 1000

w, b = np.random.randn(2)
model = lambda x: w * x + b
display(model(x))

Js = []
for i in range(n_iter):
    e = y - model(x)
    
    Js.append((e ** 2).mean())
#     if i % 10 == 0:
#         print(f'Iteration {i} - MSE: {Js[-1]}')
    if len(Js) > 1 and Js[-2] - Js[-1] < r:
        break
        
    dJ_dw = (-1 * x * e).mean()
    dJ_db = (-1 * e).mean()
    w = w - alpha * dJ_dw 
    b = b - alpha * dJ_db
    
display(model(x))
print(w, b)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()

X = x[:, np.newaxis]

model.fit(X, y)
```
